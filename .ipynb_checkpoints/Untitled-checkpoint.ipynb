# Splitting the data

```python
x=df.drop('charges',axis=1)
y=df['charges']
```

```python
from sklearn.model_selection import train_test_split as ts
```

```python
x_train,x_test,y_train,y_test=ts(x,y,test_size=0.40)
```

```python
x_test.shape,y_test.shape
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler as ss
scaler=ss()
```

```python
sc_x_train=scaler.fit_transform(x_train)
sc_x_test=scaler.fit_transform(x_test)
```

```python
pd.DataFrame(sc_x_train,columns=['age','bmi','children','Male','Smoker'])
```

# Support Vector Regression


```python
from sklearn.svm import SVR
regressor=SVR(kernel='rbf')
regressor.fit(sc_x_train,y_train)
prediction=regressor.predict(sc_x_test)
```

```python
from sklearn.metrics import mean_squared_error
cost=mean_squared_error(y_test,prediction)
cost
```

```python
sns.set_style('dark')
plt.scatter(y_test,prediction)
```
